# **Part 2: Indexing and Slicing**

```python
# Indexing and Slicing - 1D Array
import numpy as np
my_array=np.array([10,20,30,35,40,45,50,52])
print("Array ->\n",my_array)
print("First Element ->",my_array[0]) # Indexing
print("Range of Index ->\n",my_array[1:5])
```

```python
# Indexing & Slicing - 2D Array -
import numpy as np
arr_2d=np.array([[5,10,15,20,25],[10,20,30,40,50],[40,45,68,30,32]])
print("2D Array ->\n",arr_2d)
print("Indexing ->",arr_2d[1,2]) # Access element at Row-1,Column-2
print("Slicing ->\n",arr_2d[0:2,1:3]) # Row Slicier,Column Slicier
```

```python
# Indexing & Slicing - 3D Array -
arr_3d=np.array([[[5,10,15,20,25],[10,20,30,40,50]],[[7,14,21,28,35],[8,16,24,32,40]]])
print("3D Array ->\n",arr_3d)
print("Indexing ->",arr_3d[0,1,3]) # Page,Row,Column
```

# **Boolean Indexing** -

```python
# Boolean Indexing -
my_arr=np.array([5,10,15,20,25])
print("Original Array ->\n",my_arr)
# Check each element of my_arr greater than 15
arr_bool=my_arr>15
print("Boolean Array ->\n",arr_bool)
# Unmask all values from the boolean array -
print("Filtered Elements ->\n",my_arr[arr_bool])
# Direct boolean indexing -
print("Even Numbers ->\n",my_arr[my_arr%2==0])
```

## **Part 3: Reshape,Flatten,Ravel - Difference between copy and view**

```python
import numpy as np
my_arr=np.arange(1,13)
print("Main Array ->\n",my_arr)
reshape_arr=my_arr.reshape((3,4))
print("Reshapped Array ->\n",reshape_arr)
# flatten() - create a copy of the main array as 1D
flat_arr=reshape_arr.flatten()
print("Flattened Array ->\n",flat_arr)
# ravel() - create a view(if possible) of the main array as 1D
ravel_arr=reshape_arr.ravel()
print("Ravel Array ->\n",ravel_arr)
# Modify both array's
ravel_arr[0]=100
print("Modified Ravel Array ->\n",ravel_arr)
print("Original Array ->\n",reshape_arr)
flat_arr[1]=200
print("Modified Flattened Array ->\n",flat_arr)
print("Original Array ->\n",reshape_arr)
```

# **Part 4: Stacking and Splitting**

```python
# Stacking -
import numpy as np
arr_1=np.array([[5,10,15,20,25],[10,20,30,40,50]])
arr_2=np.array([[7,14,21,28,35],[12,24,36,48,60]])
print("1st Array ->\n",arr_1)
print("2nd Array ->\n",arr_2)
# Horizontal Stacking (hstack())
h_stacked=np.hstack((arr_1,arr_2)) # Join side by side
print("Horizontal Stacked Array ->\n",h_stacked)
# Vertical Stacking (vstack())
v_stacked=np.vstack((arr_1,arr_2)) # Join vertically
print("Vertical Stacked Array ->\n",v_stacked)
```

```python
# Depth Stack -
a=np.array([[1,2,3],[4,5,6]])
b=np.array([[10,20,30],[40,50,60]])
print("1st Array ->\n",a)
print("2nd Array ->\n",b)
d_stacked=np.dstack((a,b)) # Join page by page
print("Depth Stacked Array ->\n",d_stacked)
```

```python
# Splitting -
my_arr=np.arange(1,13)
print("Array ->\n",my_arr)
split_arr=np.split(my_arr,3) # Split the array in 3 parts
print("Splitted Array ->\n",split_arr)
```

```python
import numpy as np

# Assuming split_arr is the array that was previously split
# If not, you can create a sample split array for demonstration
# For this example, we'll use the 'split_arr' from the previous execution results.

# The split_arr is a list of arrays:
# print(split_arr)

joined_array = np.concatenate(split_arr)
print("Joined Array ->\n", joined_array)
```

# **Part 5: Search,Sort, and Filter of Numpy**

```python
import numpy as np
new_arr=np.array([10,20,30,40,50])
print("Array ->\n",new_arr)
# Searching -
index=np.where(new_arr==30)
print("Index where 30 is present ->",index)
indices=np.where(new_arr>30)
print("Indices where values are greater than 30 ->",indices)
```

```python
# Sorting -
my_arr=np.array([60,50,30,40,10,20])
print("Array Before Sorting ->\n",my_arr)
sort_arr=np.sort(my_arr)
print("Array After Sorting ->\n",sort_arr)
sort_desc=np.sort(my_arr)[::-1]
print("Sorted Array (descending) ->\n",sort_desc)
```

```python
# Filter in numpy -
my_arr=np.array([5,10,15,20,25,30])
print("Array ->\n",my_arr)
even_arr=my_arr[my_arr%2==0]
print("Filtered Array(Even Numbers) ->\n",even_arr)
```
